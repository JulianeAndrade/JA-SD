```python
# @title

from IPython.display import Markdown, display
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1. Introdução
display(Markdown("""
# Aplicação Prática de Python na Ciência de Dados

**Professor:** Sergio Ricardo

**Aluno:** Juliane Valente
**Objetivo:** Utilizar o Google Colab para executar comandos introdutórios em Python e analisar um dataset real utilizando bibliotecas da ciência de dados.
**Comentário:** A atividade demonstra como o uso do Python pode ser aplicado desde comandos básicos até a manipulação de dados em larga escala.
"""))

# 2. Estruturas básicas do Python
display(Markdown("""
## 1. Estruturas básicas do Python

### Loop `for` para exibir valores de 0 a 4:
"""))

for counter in range(5):
    print(counter)

display(Markdown("""
### Dicionário com frutas e quantidades:
"""))

example_dict = {'apples': 5, 'oranges': 8, 'bananas': 13}
print("Dicionário:", example_dict)

display(Markdown("""
### Conversão do dicionário em lista (apenas as chaves):
"""))
dict_to_list = list(example_dict)
print("Lista:", dict_to_list)

# 3. Carregando dados com pandas
display(Markdown("""
## 2. Carregando dados com Pandas

Vamos carregar o arquivo Excel contendo dados de inadimplência de clientes de cartão de crédito.
"""))

# Substitua o nome do arquivo abaixo se necessário após o upload no Colab
file_path = "default_of_credit_card_clients__courseware_version_1_21_19.xls"
df = pd.read_excel(file_path)

display(Markdown("### Visualizando as 5 primeiras linhas do dataset:"))
df.head()

# 4. Estrutura do dataset
display(Markdown("""
## 3. Estrutura do Dataset

Vamos verificar a dimensão do dataset e os tipos de variáveis presentes.
"""))

print("Número de linhas:", df.shape[0])
print("Número de colunas:", df.shape[1])
print("\nTipos de dados por coluna:\n")
print(df.dtypes)

colunas_numericas = df.select_dtypes(include=['number']).columns.tolist()
colunas_categoricas = df.select_dtypes(include=['object']).columns.tolist()

display(Markdown(f"""
- **Variáveis Numéricas:** {colunas_numericas}
- **Variáveis Categóricas:** {colunas_categoricas if colunas_categoricas else "Nenhuma"}
"""))

# 5. Análise estatística e dados faltantes
display(Markdown("""
## 4. Análise Estatística e Verificação de Dados Faltantes
"""))

display(df.describe())

display(Markdown("### Quantidade de valores ausentes por coluna:"))
print(df.isnull().sum())

# 6. Visualizações gráficas
display(Markdown("""
## 5. Visualizações Gráficas

### Distribuição das variáveis numéricas:
"""))

df[colunas_numericas].hist(bins=20, figsize=(16, 10))
plt.suptitle("Distribuição das Variáveis Numéricas")
plt.tight_layout()
plt.show()

display(Markdown("""
### Mapa de calor das correlações:
"""))

plt.figure(figsize=(14, 10))
sns.heatmap(df[colunas_numericas].corr(), annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Mapa de Calor da Correlação entre Variáveis Numéricas")
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.show()

display(Markdown("""
### Boxplot das variáveis numéricas (sem valores extremos):
"""))

df_limpo = df.copy()
df_limpo = df_limpo[(df[colunas_numericas] < df[colunas_numericas].quantile(0.99)).all(axis=1)]

plt.figure(figsize=(12, 10))
sns.boxplot(data=df_limpo[colunas_numericas], orient='h')
plt.title("Boxplot das Variáveis Numéricas (sem valores extremos)")
plt.xlabel("Valores")
plt.ylabel("Variáveis")
plt.tight_layout()
plt.show()
```
